```python
# 1. Setup: Install and Import
!pip install -q transformers torch

from transformers import pipeline

# 2. Load the Question Answering Pipeline
# This defaults to a DistilBERT model fine-tuned on SQuAD
qa_predictor = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# 3. Define Context and Questions
context = """
The Question Answering pipeline in Hugging Face Transformers allows you to provide a
text passage and a specific question. The model then identifies the exact span of
text where the answer is located. It provides a 'score' representing confidence,
along with 'start' and 'end' character indices for the answer within the context.
"""

questions = [
    "What does the pipeline allow you to provide?",
    "What does the score represent?",
    "Who is the current President of Mars?" # Answer not in context
]

# 4. Execution and Inspection
print(f"--- Context ---\n{context}\n")

for query in questions:
    result = qa_predictor(question=query, context=context)

    print(f"Question: {query}")
    print(f"Answer:   '{result['answer']}'")
    print(f"Score:    {round(result['score'], 4)}")
    print(f"Span:     [{result['start']}:{result['end']}]")
    print("-" * 30)

# 5. Observation: Model Behavior Under Uncertainty
# Note: Extractive QA models will often attempt to find the "best fit"
# even if the answer is missing, resulting in a low confidence score.
```
